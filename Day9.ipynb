## 라이브러리 import

```python
import requests
from bs4 import BeautifulSoup
```

## url 주소 입력 및 HTML 불러오기

```python
new_url = 'https://search.hankyung.com/apps.frm/search.news?query=코로나'
raw = requests.get(new_url)
```

## 문자열 타입의 html 코드 -> 실제 html 코드

```python
soup = BeautifulSoup(raw.text, 'html.parser')
```

## 뉴스 기사를 모두 포함하는 큰 틀 지정

* 뉴스 기사 제목 10개는 ul class = 'article' 안에 있는 모든 em class = 'tit' 태그
* 모든 데이터를 'all_title' 변수에 저장

```python
box = soup.find('ul', {'class' : 'article'})
all_title = box.find_all('em', {'class' : 'tit'})
```

## '코로나' 관련 뉴스 기사제목 10개 출력

```python
# 텍스트.strip('문자') : 텍스트 좌우에 있는 '문자' 제거

for title in all_title:
    print(title.text.strip())
```

### 첫번째 기능 추가 - 여러 페이지의 뉴스 기사 제목 수집

* url 주소 패턴 : url&page=(number) 

```python
for page in range(1, 11):
    new_url = 'https://search.hankyung.com/apps.frm/search.news?query=코로나' + str(page)
    raw = requests.get(new_url)
    
    soup = BeautifulSoup(raw.text, 'html.parser')
    
    box = soup.find('ul', {'class' : 'article'})
    all_title = box.find_all('em', {'class' : 'tit'})
    
    print('<' + str(page) + '페이지 뉴스 기사 제목' + '>')
    
    for title in all_title:
        print(title.text.strip())
    
    print()
```

### 두번째 기능 추가 - input( )으로 키워드 입력받기

* '코로나' 검색 결과 주소 안에 '코로나'라는 키워드 포함되어 있음
* input을 통해 직접 키워드 입력 가능

```python
keyword = input('뉴스 검색 키워드: ')
count = 0

for page in range(1, 11):
    new_url = 'https://search.hankyung.com/apps.frm/search.news?query=' + keyword + '&page' + str(page)
    raw = requests.get(new_url)
    
    soup = BeautifulSoup(raw.text, 'html.parser')
    
    box = soup.find('ul', {'class' : 'article'})
    all_title = box.find_all('em', {'class' : 'tit'})
    
    for title in all_title:
        count += 1
        print(count, '-', title.text.strip())
```

### 9일차 출석미션

* [날짜 + 시간] + 뉴스 기사 제목

```python
keyword = input('뉴스 검색 키워드: ')
count = 0

for page in range(1, 3):
    new_url = 'https://search.hankyung.com/apps.frm/search.news?query=' + keyword + '&page' + str(page)
    raw = requests.get(new_url)
    
    soup = BeautifulSoup(raw.text, 'html.parser')
    
    box = soup.find('ul', {'class' : 'article'})
    all_title = box.find_all('em', {'class' : 'tit'})

    date = soup.find_all('span', {'class' : 'date_time'})
            
    for date, title in zip(date, all_title):
        count += 1
        print(count, '-', '[', date.text, ']', title.text.strip())
        
```
